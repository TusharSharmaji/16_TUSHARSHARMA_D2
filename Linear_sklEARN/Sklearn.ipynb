```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
df= pd.read_csv("/content/study_hours_cgpa_200_rows.csv")
```

```python
df.head()
```

```python
plt.scatter(df['StudyHours'],df['CGPA'])
plt.xlabel('StudyHours')
plt.ylabel('CGPA')
```

```python
X = df.iloc[:,0:1]
y = df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train.values.reshape(-1, 1), y_train)
```

```python
X_test
```

```python
lr.predict(X_test.iloc[[0]].values.reshape(1,1))
```

```python
plt.scatter(df['StudyHours'],df['CGPA'])
plt.plot(X_train,lr.predict(X_train.values.reshape(-1, 1)),color='black')
plt.xlabel('StudyHours')
plt.ylabel('CGPA')
```

```python
m = lr.coef_
m
```

```python
b = lr.intercept_
b
```
