```python
def pallindrom(n):
  original=n
  rev=0

  while(n>0):
    rem=n%10
    rev=rev*10+rem
    n=n//10

  if(original==rev):
    print("Number is pallindrom")
  else:
    print("Number is not pallindrom")

n=int(input("Enter the number"))
pallindrom(n)
```

```python
l=[1,2,3,4,5]

# Handle edge cases for empty or single-element lists
if len(l) < 2:
    print("List must contain at least two elements to find a second maximum.")
else:
    max1 = float('-inf') # Initialize first maximum to negative infinity
    max2 = float('-inf') # Initialize second maximum to negative infinity

    for num in l:
        if num > max1:
            max2 = max1 # Current max1 becomes new max2
            max1 = num  # Current num becomes new max1
        elif num > max2 and num != max1: # If num is smaller than max1 but greater than max2
            max2 = num

    if max2 == float('-inf'):
        print("No distinct second maximum found (e.g., all elements are the same).")
    else:
        print(f"The second largest element is: {max2}")
```

```python
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"
    elif n == 0:
        return 1
    else:
        fact = 1
        for i in range(1, n + 1):
            fact = fact * i
        return fact

# Example usage:
num = 5
print(f"The factorial of {num} is {factorial(num)}")

num = 0
print(f"The factorial of {num} is {factorial(num)}")

um = -3
print(f"The factorial of {num} is {factorial(num)}")
```
